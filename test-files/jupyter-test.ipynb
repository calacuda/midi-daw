```python
def start_event():
    wait_for("1")
```

```python
def sp(pad_n: int) -> int:
    if pad_n <= 4:
        return 47 + pad_n
    elif pad_n <= 8:
        return 43 + (pad_n - 4)
    elif pad_n <= 12:
        return 39 + (pad_n - 8)
    elif pad_n <= 16:
        return 35 + (pad_n - 12)
```

```python
def do_drums(pad):
    note([sp(4), sp(1)], sn(), vel=100)
    note(sp(1), sn(1), vel=100)
    note(
        [sp(7), sp(1)],
        sn(),
        vel=127,
    )
    note(sp(1), sn(1), vel=100)
    note([sp(4), sp(1)], sn(), vel=100)
    note(sp(1), sn(1), vel=100)
    note([sp(7)], sn(), vel=100)
    note(sp(pad), sn(1), vel=100)
```

```python
@play_on("sp404", channel=Ch1, block=False, loop=-1, setup=start_event)
def drums():
    # 36 - 52
    do_drums(1)
    do_drums(9)
    
```

```python
drums()
```

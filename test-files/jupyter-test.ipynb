````python
def start_event():
    wait_for("1")
````

````python
def sp(pad_n: int) -> int:
    if pad_n <= 4:
        return 47 + pad_n
    elif pad_n <= 8:
        return 43 + (pad_n - 4)
    elif pad_n <= 12:
        return 39 + (pad_n - 8)
    elif pad_n <= 16:
        return 35 + (pad_n - 12)
````

````python
def start_on(beat):
    def _start_on():
        return wait_for(beat)

    return _start_on
````

````python
# ```python

times_looped = 0

@play_on("sp404", channel=Ch1, block=False, loop=-1, setup=start_event)
def four_on_floor(vel):
    global times_looped
    clap_beat = [sp(2), sp(4), sp(9)] if times_looped >= 4 else [sp(2), sp(4)] 
    
    for n in (
        [sp(2), sp(4)],
        clap_beat,
        [sp(2), sp(4)],
        clap_beat,
        
        # [sp(2), sp(4), sp(9)],
    ):
        if times_looped >= 12:
            n = n + [sp(10)]
        note(n, en(), vel)
        rest(sn())
        
        if times_looped >= 8:
            note(sp(10), en(), vel)
        else:
            rest(en())
        
        rest(sn())
        

    if times_looped < 16:
        times_looped += 1
    
four_on_floor.stop()
four_on_floor(100)
# four_on_floor.stop()
# ```
````

````python
@play_on("micro", channel=Ch1, block=False, loop=-1, setup=start_event)
def chords(finish_event):
    vel = 60
    o = 4

    def dm(n_dur):
        note([f"d{o}", f"f{o}", f"a{o}"], n_dur, vel)

    def gm_inv(n_dur):
        note([f"d{o}", f"g{o}", f"b{o}"], n_dur, vel)

    def cm(n_dur):
        note([f"c{o}", f"e{o}", f"g{o}"], n_dur, vel)

    def fm_inv(n_dur):
        note([f"c{o}", f"f{o}", f"a{o}"], n_dur, vel)

    dm(qn())
    rest(sn())
    dm(qn())
    rest(sn())
    dm(qn())
    rest(sn())
    dm(qn())
    # rest(sn())
    gm_inv(qn())
    rest(tn())

    dm(qn())
    rest(sn())
    dm(qn())
    rest(sn())
    dm(qn(2))
    # rest(sn())
    cm(hn())
    rest(tn())

    trigger(finish_event)


# chords.stop()
# panic(chords.new_midi_out)
# chords()
````

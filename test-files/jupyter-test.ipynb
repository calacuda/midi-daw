````python
def start_event():
    wait_for("1")
````

````python
def sp(pad_n: int) -> int:
    if pad_n <= 4:
        return 47 + pad_n
    elif pad_n <= 8:
        return 43 + (pad_n - 4)
    elif pad_n <= 12:
        return 39 + (pad_n - 8)
    elif pad_n <= 16:
        return 35 + (pad_n - 12)
````

````python
def do_drums(pad):
    note([sp(4), sp(1)], sn(), vel=100)
    note(sp(1), sn(1), vel=100)
    note(
        [sp(7), sp(1)],
        sn(),
        vel=127,
    )
    note(sp(1), sn(1), vel=100)
    
    note([sp(4), sp(1)], sn(), vel=100)
    note(sp(11), sn(1), vel=100)
    note([sp(7)], sn(), vel=100)
    note(sp(pad), sn(1), vel=100)
````

````python
@play_on("sp404", channel=Ch1, block=False, loop=-1, setup=start_event)
def drums():
    # 36 - 52
    do_drums(1)
    do_drums(9)
    
````

````python
def start_on(beat):
    def _start_on():
        return wait_for(beat)

    return _start_on
````

````python
def f_on_f_stop():
    _four_on_floor_beat.stop()
    _four_on_floor_clap.stop()
    _four_on_floor_hh.stop()


@play_on("sp404", channel=Ch1, block=False, loop=-1, setup=start_event)
def _four_on_floor_beat(step, vel):
    note([sp(2), sp(4)], step(), vel)
    rest(step(3))


@play_on("sp404", channel=Ch1, block=False, loop=-1)
def _four_on_floor_hh(step, vel):
    rest(step(2))
    note(sp(1), step(), vel)
    rest(step())


@play_on("sp404", channel=Ch1, block=False, loop=-1)
def _four_on_floor_clap(step, vel):
    rest(step(4))
    note(sp(9), step(), vel)
    rest(step(3))


# @play_on("sp404", channel=Ch1, block=False, loop=1, setup=start_event)
def four_on_floor():
    vel = 100
    step = en

    # wait_for("1")
    log.info("start main beat")
    _four_on_floor_beat(step, vel) 
    start_event()
    start_event()
    log.info("start clap")
    _four_on_floor_clap(step, vel)
    # wait_for("1")
    log.info("start hh")
    _four_on_floor_hh(step, vel)


f_on_f_stop()
four_on_floor()
````
